Nhập vào một dòng gồm các số nguyên. Hãy loại bỏ phần tử trùng nhưng 
giữ nguyên thứ tự xuất hiện đầu tiên, rồi in ra list kết quả. 
Yêu cầu in ra đúng dạng list Python, ví dụ [1, 2, 3]

```python
input_data = input().split()
numbers = [int(x) for x in input_data]

def remove_duplicates(input_list):
    seen = set()
    result = []
    for item in input_list:
        if item not in seen:
            result.append(item)
            seen.add(item)
    return result
final_list = remove_duplicates(numbers)
print(final_list)
```

Nhập một dòng số nguyên. In ra list tổng tích lũy (phần tử i là tổng từ đầu 
tới vị trí i), giữ nguyên thứ tự

```python
nums = list(map(int, input().split()))

prefix_sum = []
current_total = 0

for x in nums:
    current_total += x
    prefix_sum.append(current_total)

print(prefix_sum)
```

Nhập một tuple số nguyên (một dòng) và số nguyên k (dòng kế). In ra tuple 
sau khi xoay trái k vị trí. (Giữ nguyên thứ tự tương đối sau xoay; không sắp 
xếp.)

```python
tup = tuple(map(int, input().split()))
k = int(input())

if len(tup) > 0:
    k = k % len(tup)
    result = tup[k:] + tup[:k]
else:
    result = tup

print(result)
```

Nhập một dòng gồm các cặp key:value cách nhau bởi dấu cách. Tạo 
dictionary ánh xạ mỗi key -> list các value theo thứ tự xuất hiện. Nếu key 
lặp lại, thêm vào list của key đó

```python
data_input = input().split()

result_dict = {}

for item in data_input:
    key, value = item.split(':')
    if key not in result_dict:
        result_dict[key] = []
    
    result_dict[key].append(value)

print(result_dict)
```

Nhập một list số nguyên, In ra dictionary {'positives': ..., 'negatives': ..., 
'zeros': ...} (đếm số >0, <0, =0)

```python
nums = list(map(int, input().split()))
counts = {'positives': 0, 'negatives': 0, 'zeros': 0}

for x in nums:
    if x > 0:
        counts['positives'] += 1
    elif x < 0:
        counts['negatives'] += 1
    else:
        counts['zeros'] += 1

print(counts)
```

Viết chương trình nhập vào một list các chuỗi, sau đó đếm số lần xuất hiện 
của từng chuỗi và lưu vào dictionary.

```python
strings = input().split()

count_dict = {}

for s in strings:
    if s in count_dict:
        count_dict[s] += 1
    else:
        count_dict[s] = 1

print(count_dict)
```

Viết chương trình nhập vào hai dictionary. Sau đó gộp lại thành một 
dictionary mới trong đó nếu một key xuất hiện trong cả hai dict, giá trị sẽ 
được cộng lại, nếu chỉ có trong một dict thì giữ nguyên. In ra dictionary mới 
theo thứ tự từ điển tăng dần của key.

```python

def input_to_dict(prompt):
    print(prompt)
    line = input().split()
    d = {}
    for item in line:
        key, value = item.split(':')
        d[key] = int(value)
    return d
dict1 = input_to_dict("Nhập dict 1 (định dạng key:value cách nhau bởi dấu cách):")
dict2 = input_to_dict("Nhập dict 2 (định dạng key:value cách nhau bởi dấu cách):")

result_dict = dict1.copy()


for key, value in dict2.items():
    if key in result_dict:
        result_dict[key] += value
    else:
        result_dict[key] = value

sorted_result = dict(sorted(result_dict.items()))

print("\nDictionary sau khi gộp và sắp xếp:")
print(sorted_result)
```

Viết chương trình nhập vào một danh sách các số nguyên, và số nguyên 
dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là 
một tuple, hai số trong tuple có thể giống nhau, thứ tự xuất hiện trong danh 
sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple trong list được 
sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple

```python

try:
    nums = list(map(int, input("Nhập danh sách số nguyên (cách nhau bởi dấu cách): ").split()))
    k = int(input("Nhập số nguyên dương k: "))
except ValueError:
    print("Vui lòng nhập đúng định dạng số!")
    exit()

pairs = []
n = len(nums)
for i in range(n):
    for j in range(i + 1, n):
        if nums[i] + nums[j] == k:
            pairs.append((nums[i], nums[j]))
sorted_pairs = sorted(pairs, key=lambda x: x[0])
print(sorted_pairs)
```

[Tách số chẵn và lẻ từ tuple] 
Nhập vào một tuple chứa các số nguyên. Hãy tạo ra hai tuple mới: một chứa 
tất cả các số chẵn và một chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên 
thứ tự tương đối. In ra hai tuple này trên hai dòng riêng biệt.

```python

try:
    input_data = input("Nhập các số nguyên (cách nhau bởi dấu cách): ").split()
    original_tuple = tuple(map(int, input_data))

    even_tuple = tuple(x for x in original_tuple if x % 2 == 0)

    odd_tuple = tuple(x for x in original_tuple if x % 2 != 0)

    print(even_tuple)
    print(odd_tuple)

except ValueError:
    print("Vui lòng chỉ nhập các số nguyên!")
```

[Tìm phần tử xuất hiện nhiều nhất] 
Nhập vào một list các số nguyên. Hãy tìm số xuất hiện nhiều lần nhất trong 
list. Nếu có nhiều số cùng có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ 
nhất trong số đó

```python
from collections import Counter
nums = list(map(int, input("Nhập dãy số: ").split()))

if not nums:
    print("Danh sách rỗng")
else:
    counts = Counter(nums)
    max_freq = max(counts.values())
    most_common_nums = [num for num, freq in counts.items() if freq == max_freq]
    print(min(most_common_nums))
```

[Đảo ngược key và value của dictionary] 
Nhập vào một dictionary có value là duy nhất (không trùng nhau). Hãy tạo 
một dictionary mới bằng cách đảo ngược key và value của dictionary ban 
đầu. In ra dictionary mới.

```python
line = input("Nhập dictionary (key:value): ").split()
original_dict = {}

for item in line:
    key, value = item.split(':')
    original_dict[key] = value
reversed_dict = {v: k for k, v in original_dict.items()}

print(reversed_dict)
```

[Lọc dictionary theo giá trị] 
Nhập vào một dictionary có key là chuỗi và value là số nguyên, và một số 
nguyên k ở dòng tiếp theo. Hãy tạo một dictionary mới chỉ chứa các cặp 
key-value từ dictionary ban đầu mà có value lớn hơn k. In ra dictionary 
mới.

```python
data_input = input().split()
d = {}
for item in data_input:
    key, value = item.split(':')
    d[key] = int(value)

k = int(input())
filtered_dict = {key: value for key, value in d.items() if value > k}
print(filtered_dict)
```
