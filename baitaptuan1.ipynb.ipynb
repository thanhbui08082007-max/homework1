```python
#Bài 1
# Bước 1: Nhập dữ liệu và ép kiểu sang số nguyên (int)
n = int(input("Nhập vào số nguyên n: "))

# Bước 2: Tính số gấp đôi
ket_qua = n * 2

# Bước 3: In kết quả ra màn hình
print("Số gấp đôi n là:", ket_qua)
```

```python
#Bài 2
# Nhập dữ liệu
a = float(input("Nhập chiều rộng a: "))
b = float(input("Nhập chiều dài b: "))

# Tính toán
r = a / 2
s_hcn = a * b
s_ht = 3.14 * (r ** 2)
s_cl = s_hcn - s_ht

# Xuất kết quả làm tròn 2 chữ số thập phân
print("%.2f" % s_cl)
```

```python
#Bài 3
c = input("Nhập vào một ký tự: ")

# Kiểm tra nếu độ dài nhập vào đúng 1 ký tự
if len(c) == 1:
    # Nếu là chữ HOA (ASCII từ 65 đến 90)
    if 'A' <= c <= 'Z':
        ket_qua = chr(ord(c) + 32)
        print(f"Kết quả: {ket_qua}")
    
    # Nếu là chữ thường (ASCII từ 97 đến 122)
    elif 'a' <= c <= 'z':
        ket_qua = chr(ord(c) - 32)
        print(f"Kết quả: {ket_qua}")
    
    else:
        print("Đây không phải là chữ cái!")
else:
    print("Vui lòng chỉ nhập đúng 1 ký tự.")
```

```python
#Bài 4
c = input("Nhập vào một kí tự: ")

# Kiểm tra nếu người dùng nhập đúng 1 ký tự
if len(c) == 1:
    # Điều kiện: nằm trong khoảng a-z HOẶC nằm trong khoảng A-Z
    if ('a' <= c <= 'z') or ('A' <= c <= 'Z'):
        print(f"'{c}' là kí tự alphabet")
    else:
        print(f"'{c}' không phải là kí tự alphabet")
else:
    print("Vui lòng chỉ nhập duy nhất 1 kí tự!")
```

```python
#Bài 5
c = input("Nhập vào một chữ cái hoa: ")

# Kiểm tra điều kiện đầu vào là chữ hoa
if 'A' <= c <= 'Z':
    # Bước 1: Chuyển về chữ thường tương ứng
    chu_thuong = c.lower()
    
    # Bước 2: Xét trường hợp đặc biệt 'A'
    if c == 'A':
        ket_qua = 'z'
    else:
        # Bước 3: Lấy ký tự liền trước bằng cách trừ đi 1 trong mã ASCII
        ket_qua = chr(ord(chu_thuong) - 1)
        
    print(f"Ký tự liền trước của '{chu_thuong}' là: {ket_qua}")
else:
    print("Vui lòng nhập một chữ cái viết HOA!")
```

```python
#Bài 6
import math

# Nhập dữ liệu và chuyển sang số thực (float)
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))

# 1. Kiểm tra điều kiện tam giác
if a > 0 and b > 0 and c > 0 and (a + b > c) and (a + c > b) and (b + c > a):
    # 2. Tính nửa chu vi
    p = (a + b + c) / 2
    
    # 3. Tính diện tích theo công thức Heron
    s = math.sqrt(p * (p - a) * (p - b) * (p - c))
    
    # 4. In kết quả làm tròn 1 chữ số thập phân
    print(f"Dien tich tam giac la: {s:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
#Bài 7
# Nhập chuỗi từ bàn phím
s = input("Nhập vào một chuỗi (độ dài >= 20): ")

# Kiểm tra độ dài chuỗi trước khi truy xuất để tránh lỗi
if len(s) >= 20:
    # Chữ cái thứ năm (index 4)
    chu_thu_5 = s[4]
    
    # Chữ cái thứ chín (index 8)
    chu_thu_9 = s[8]
    
    print(f"Chữ cái thứ năm là: {chu_thu_5}")
    print(f"Chữ cái thứ chín là: {chu_thu_9}")
else:
    print("Chuỗi quá ngắn! Vui lòng nhập chuỗi có ít nhất 20 ký tự.")
```

```python
#Bài 8
name = input("Ten chu ho: ")
cu = int(input("Chi so thang truoc: "))
moi = int(input("Chi so thang nay: "))

# Tính số điện tiêu thụ
kwh = moi - cu

# Khởi tạo biến tổng tiền
tong_tien = 0

if kwh > 400:
    tong_tien += (kwh - 400) * 3460
    kwh = 400
if kwh > 300:
    tong_tien += (kwh - 300) * 3350
    kwh = 300
if kwh > 200:
    tong_tien += (kwh - 200) * 2998
    kwh = 200
if kwh > 100:
    tong_tien += (kwh - 100) * 2380
    kwh = 100
if kwh > 50:
    tong_tien += (kwh - 50) * 2050
    kwh = 50
if kwh > 0:
    tong_tien += kwh * 1984

# Tính thêm VAT 8% và làm tròn đến đơn vị đồng
tien_cuoi = round(tong_tien * 1.08)

print("--- Kết quả ---")
print(f"Ho va ten: {name}")
print(f"Tien phai tra la: {tien_cuoi}")
```
