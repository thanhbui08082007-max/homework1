Viết chương trình Python sử dụng một trong 2 loại vòng lặp for hoặc while để 
tính tống các số nguyên từ 1 đến n. (n <= 1000)

```python
n = int(input())
if n > 1000:
    print("Vui lòng nhập n nhỏ hơn hoặc bằng 1000.")
else:
    tong = 0
    for i in range(1, n + 1):
        tong = tong + i
    
    print(f"Tổng các số từ 1 đến {n} là: {tong}")
```

VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi. 
Kiểm tra số này có phải là số nguyên tố hay không?

```python
import math

n = 0
while n <= 0:
    n = int(input("Nhập vào một số nguyên dương (n > 0): "))
    if n <= 0:
        print("Số vừa nhập không phải số nguyên dương")

so_nguyen_to = True

# Số nhỏ hơn 2 không phải là số nguyên tố
if n < 2:
    so_nguyen_to = False
else:
    i = 2
    while i <= math.sqrt(n):
        if n % i == 0:
            so_nguyen_to = False
            break
        i += 1

if so_nguyen_to:
    print(f"{n} là số nguyên tố.")
else:
    print(f"{n} không phải là số nguyên tố.")
```

Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn 
[a, b]. Mỗi dòng là một bộ test (a b)."

```python
import math

a = 0
b = 0
while True:
    line = input()
    parts = line.split()
    
    if len(parts) == 2:
        a = int(parts[0])
        b = int(parts[1])
        if a > 0 and b > 0 and a <= b:
            break
    print("Vui lòng nhập lại")

tong_nguyen_to = 0

for n in range(a, b + 1):
    if n < 2:
        continue
    
    is_prime = True
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        tong_nguyen_to += n
print(f"Tổng các số nguyên tố trong đoạn [{a}, {b}] là: {tong_nguyen_to}")
```

Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số  đối xứng (palindrome). In ra số bước và giá trị palindrome thu được.

```python
n = int(input("Nhập số nguyên dương n: "))
steps = 0

while str(n) != str(n)[::-1]:
    dao_nguoc = int(str(n)[::-1])
    
    n = n + dao_nguoc
    steps += 1

print("Số bước:", steps)
print("Số đối xứng thu được:", n)
```

Nhập vào số nguyên dương n. In ra tất cả các số chính phương ≤ n có các chữ số 
đôi một khác nhau (in theo thứ tự tăng dần, cùng hàng, cách nhau bởi một dấu 
cách). Nếu không có, in ra (no number).

```python
import math
n = int(input())

found = False 
first = True

i = 1
while i * i <= n:
    square = i * i
    s_square = str(square)
    
    if len(s_square) == len(set(s_square)):
        if not first:
            print(" ", end="")
        print(square, end="")
        found = True
        first = False
    
    i += 1

if not found:
    print("(no number)")
else:
    print()
```

Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất 
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất.

```python
n = int(input())

max_length = 0
best_x = 0

for i in range(1, n + 1):
    current = i
    length = 1
    
    temp = current
    while temp != 1:
        if temp % 2 == 0:
            temp = temp // 2
        else:
            temp = 3 * temp + 1
        length += 1
    if length > max_length:
        max_length = length
        best_x = i

print(best_x, max_length)
```

1 Viết chương trình đếm số lượng ước số chẵn của số nguyên dương n với n < 10^6.

```python
import math

n = int(input())
count = 0
for i in range(1, int(math.sqrt(n)) + 1):
    if n % i == 0:
        if i % 2 == 0:
            count += 1

        if (n // i) != i:
            if (n // i) % 2 == 0:
                count += 1

print(count)
```

vừa đúng bằng 220 
Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. 
Nếu có, in ra màn hình "true", ngược lại, in ra "false"

```python
a= int(input())
b = int(input())
sum_a = 0
sum_b = 0
if a > 1:
    sum_a = 1
    i = 2
    while i * i <= a:
        if a%i == 0:
            if i * i == a:
                sum_a += i
            else:
                sum_a += i + (a//i)
        i += 1
if b > 1:
    sum_b = 1
    j = 2
    while j * j <= b:
        if b%j ==0:
            if j * j == b:
                sum_b += j
            else:
                sum_b += j + (b//j)
        j +=1
if sum_a == b or sum_b == a:
    print("true")
else:
    print("false")

```
