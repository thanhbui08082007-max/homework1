```python
a, b, c = map(float,input().split())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("day la tam giac deu")
    elif a == b or b == c or a == c:
        print("day la tam giac can")
    elif a ** 2 + b ** 2 == c ** 2 or a ** 2 + c ** 2 == b ** 2 or b ** 2 + c ** 2 == a ** 2:
        print("day la tam giac vuong")
    else:
        print("day la tam giac thuong")
    p = (a + b + c) / 2
    d = p * 2
    s = (p * (p - a) * (p - b) * (p - c)) ** 0.5
    print(f"chu vi la {d} dien tich la {s}")
else:
    print("day khong phai tam giac")    

```

Bài 8:Giải hệ phương trình tuyến tính:
ax+by=m
cx+dy=n

```python
a, b, c, d, m, n = map(int,input().split())
D = a * d - b * c
Dx = m * d - n * b
Dy = a * n - c * m

if D != 0:
    print("Phương trình có nghiệm")
    x = Dx/D
    y = Dy/D
    print(f"x là {x} , y là {y}")
else:
    print("Phương trình vô nghiệm")
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vào một
điểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
#Tâm I(a,b) của đường tròn
a, b = map(float,input().split())
x, y = map(float,input().split())
r = float(input())
khoang_cach = ((x-a)**2 + (y-b)**2)**0.5
if khoang_cach < r:
    print("Điểm A nằm trong đường tròn")
elif khoang_cach == r:
    print("Điểm A thuộc đường tròn")
else:
    print("Điểm A nằm ngoài đường tròn")
```
